```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import figure
```

```python
zID = [1, 8, 7, 5, 1, 10, 1]

A = zID[1]
B = zID[2]
C = zID[3]
D = zID[4]
E = zID[5]
F = zID[6]
```

```python
arrSize = 10000
normal = arrSize / 100
xInc = np.array([])
for i in range(0,arrSize):
    xInc = np.append(xInc,i/normal)
```

```python
Gt = -F*np.exp(-np.absolute((xInc/E)-C))*np.cos(((2*np.pi)/E)*(xInc-(B/10)))-A
GtPrime = np.gradient(Gt)
GtPrime2 = np.gradient(GtPrime)
```

```python
yMax = -A
i = 0
while i < xInc.shape[0]:
    i += 1
    if Gt[i-1] > yMax:
        yMax = Gt[i-1]

deltaY = yMax + A
maxVal = deltaY * .44 - A
minVal =  -A - deltaY * .44


figure(figsize=(9,9), dpi=80)

for i in range(0,arrSize):
    if GtPrime2[i] >= 0 and Gt[i] > maxVal:
        plt.plot(xInc[i:i+2],Gt[i:i+2],'b')
    elif GtPrime2[i] >= 0 and Gt[i] < minVal:
        plt.plot(xInc[i:i+2],Gt[i:i+2],'b')
    else:
        plt.plot(xInc[i:i+2],Gt[i:i+2],'gray')
```
