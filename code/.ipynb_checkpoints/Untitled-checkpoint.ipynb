```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import figure
```

```python
zID = [1, 5, 3, 8, 10, 7, 8]

A = zID[1]
B = zID[2]
C = zID[3]
D = zID[4]
E = zID[5]
F = zID[6]
```

```python
xInc = np.array([])
for i in range(0,1001):
    xInc = np.append(xInc,i/10)
```

```python
Gt = -F*np.exp(-np.absolute((xInc/E)-C))*np.cos(((2*np.pi)/E)*(xInc-(B/10)))-A
GtPrime = np.gradient(Gt)
GtPrime2 = np.gradient(GtPrime)
```

```python
iAmTooLazy = -4.9438
figure(figsize=(9,9), dpi=80)

for i in range(0,1000):
    if GtPrime2[i] >= 0 and Gt[i] > iAmTooLazy:
        plt.plot(xInc[i:i+2],Gt[i:i+2],'b')
    else:
        plt.plot(xInc[i:i+2],Gt[i:i+2],'gray')
```
